## Data Understanding 

df1 table
Column Name:Description
> online_order :Indicates if online ordering is available (Yes/No)

> book_table:Indicates if table booking is available (Yes/No)

> rate:Average rating (e.g., 4.1/5)

> votes:Number of votes received by the restaurant

> rest_type:Type of restaurant (e.g., Casual Dining, Quick Bites)

> dish_liked:Popular dishes liked by customers

> cuisines:Cuisines offered by the restaurant

> approx_costfor_two_people:Approximate cost for two people in INR

> listed_intype:Restaurant listing category (e.g., Buffet, Delivery)

> listed_incity:Local area in Bangalore where the restaurant is located



df2 table Column Name : Description
> listed_incity:Local area name (used for merging)
 
>Latitude:Latitude of the location

>Longitude:Longitude of the location

```python
from folium.plugins import MarkerCluster
import folium
from IPython.display import IFrame
```

```python
italian_df=df_m[df_m['cuisines'].str.contains('Italian',na=False)]
map4=folium.Map(location=[12.9716,77.5946],zoom_start=12)
for i , row in italian_df.iterrows():
    folium.Marker(
        location=[row['Latitude'],row['Longitude']],
        popup=f"{row['listed_incity']}<br>Cuisine: {row['cuisines']}",
        icon=folium.Icon(color='purple')
    ).add_to(map4)
map4.save("italian_spots_map.html")
```

```python
IFrame('italian_spots_map.html',width=1000,height=800)
```

## Interactive Restaurant Density Mapping using Folium with Marker Clustering 

```python
map1 = folium.Map(location=[12.9716,77.5946], zoom_start=12)

marker_cluster=MarkerCluster().add_to(map1)

for i , row in df_m.iterrows():
    if not pd.isna(row['Latitude']) and not pd.isna(row['Longitude']):
        folium.Marker(
            location=[row['Latitude'],row['Longitude']],
                      popup=f"{row['listed_incity']}<br>Rating:{row['rate']}<br>Cost: {row['approx_costfor_two_people']}",
                      ).add_to(marker_cluster)
map1.save('restaurant_density.html')                     
```

```python
IFrame('restaurant_density.html',width=1000,height=800)
```
